# Lao–English Text Analysis Project

**Objective:**  
This project analyzes Lao and English text data using basic NLP techniques, 
including text cleaning, tokenization, word frequency analysis, and comparison 
between the two languages.

**Tools:** Python, NLTK, spaCy, Pandas, Matplotlib


```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def load_text(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return f.read()

lao_text = load_text("data/lao_corpus.txt")
eng_text = load_text("data/english_corpus.txt")

print("Lao text length:", len(lao_text))
print("English text length:", len(eng_text))
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

lao_clean = clean_text(lao_text)
eng_clean = clean_text(eng_text)
```

```python
lao_tokens = lao_clean.split()
eng_tokens = eng_clean.split()

print("Lao tokens:", len(lao_tokens))
print("English tokens:", len(eng_tokens))
```

```python
from collections import Counter

lao_freq = Counter(lao_tokens)
eng_freq = Counter(eng_tokens)

print("Top 10 Lao words:")
print(lao_freq.most_common(10))

print("\nTop 10 English words:")
print(eng_freq.most_common(10))
```

```python
lao_df = pd.DataFrame(lao_freq.most_common(20), columns=["Word", "Frequency"])
eng_df = pd.DataFrame(eng_freq.most_common(20), columns=["Word", "Frequency"])

lao_df, eng_df
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(eng_df["Word"], eng_df["Frequency"])
plt.xticks(rotation=45)
plt.title("Top English Words")
plt.show()

plt.figure()
plt.bar(lao_df["Word"], lao_df["Frequency"])
plt.xticks(rotation=45)
plt.title("Top Lao Words")
plt.show()
```

## Results & Discussion

- English text shows higher repetition of common function words.
- Lao text demonstrates different token distribution due to language structure.
- The results highlight linguistic differences between Lao and English text data.
